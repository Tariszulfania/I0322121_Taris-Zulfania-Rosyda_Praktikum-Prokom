```python
#Buatlah sebuah fungsi untuk mencari invers matrix berdasarkan input matrix dari user

def invers():
    det = []

    #Tulis kode kamu disini
def invers():
    baris = int(input('masukan jumlah baris: '))
    kolom = int(input('masukan jumlah kolom: '))

    matriks= create_matriks(baris, kolom)
    det= determinan(matriks)

    if det == 0:
        print('matriks tidak ada invers karena determinan sama dengan 0')
    else:
        inverse_matriks= inverse(matriks, det)
        print()
        print_matriks(inverse_matriks)

def create_matriks(baris, kolom):
    matriks = []
    for i in range(baris):
        row = []
        for j in range(kolom):
            element = float(input(f'Masukkan elemen matriks pada baris ke-{i+1}, kolom ke-{j+1}: '))
            row.append(element)
        matriks.append(row)
    return matriks

def determinan(matriks):
    n = len(matriks)
    if n != len(matriks[0]):
        raise ValueError('Matriks memiliki bentuk persegi, jumlah baris sama dengan jumlah kolom)')
    
    if n == 1:
        return matriks[0][0]
    
    det = 0
    for j in range(n):
        submatrix = []
        for i in range(1, n):
            row = matriks[i][:j] + matriks[i][j+1:]
            submatrix.append(row)
        det += (-1) ** j * matriks[0][j] * determinan(submatrix)
    
    return det

def cofactor(matriks):
    n = len(matriks)
    cofactor_matriks = []
    for i in range(n):
        row = []
        for j in range(n):
            submatrix = []
            for k in range(n):
                if k != i:
                    subrow = matriks[k][:j] + matriks[k][j+1:]
                    submatrix.append(subrow)
            sign = (-1) ** (i + j)
            cofactor_ij = sign * determinan(submatrix)
            row.append(cofactor_ij)
        cofactor_matriks.append(row)
    return cofactor_matriks

def transpose(matriks):
    n = len(matriks)
    transposed_matriks = []
    for j in range(n):
        row = []
        for i in range(n):
            row.append(matriks[i][j])
        transposed_matriks.append(row)
    return transposed_matriks

def inverse(matriks, det):
    n = len(matriks)
    cofactor_matriks = cofactor(matriks)
    adjugate_matriks = transpose(cofactor_matriks)
    inverse_matriks = []
    for i in range(n):
        row = []
        for j in range(n):
            element = adjugate_matriks[i][j] / det
            row.append(element)
        inverse_matriks.append(row)
    return inverse_matriks

def print_matriks(matriks):
    for row in matriks:
        print(row)


invers()
print()
```
